```python
import pandas as pd
```

```python
# Get Dataframe from URL
url  = 'https://raw.githubusercontent.com/waldronlab/curatedMetagenomicDataCuration/master/inst/curated/LokmerA_2019/LokmerA_2019_metadata.tsv'
df = pd.read_csv(url, sep='\t')
df.head()
```

```python
keep_columns = ['NCBI_accession', 'gender', 'BMI', 'lifestyle', 'population']
df_filtered = df[keep_columns]
df_filtered.head()
```

```python
# print all NCBI_accessions
print(df_filtered['NCBI_accession'].to_string(index=False))
```
